```python
import pandas as pd
import numpy as np
from time import time
```

```python
def preprocess_data(df):
    """Preprocess the entire dataset."""
    df = df.dropna()
    numeric_cols = df.select_dtypes(include='number').columns
    df[numeric_cols] = (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()
    df['Mean'] = df[numeric_cols].mean(axis=1)

    return df


def sequential_process(file_path):
    """Preprocess data without parallelism."""
    start_time = time()
    df = pd.read_csv(file_path)
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
    df = df.dropna(subset=['date'])  # Drop rows with invalid dates

    # Preprocess the data
    processed_df = preprocess_data(df)

    # Save the processed data
    processed_df.to_csv("/content/drive/MyDrive/P&DC /Theory/project/processed_sequential.csv", index=False)
    print(f"Sequential processing completed in {time() - start_time:.2f} seconds")

if __name__ == "__main__":
    # File path
    input_file = "/content/drive/MyDrive/P&DC /Theory/project/data.csv"  # Replace with your actual file path

    # Sequential processing
    print("Running sequential processing...")
    sequential_process(input_file)
```
